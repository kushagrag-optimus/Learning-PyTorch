# **Numpy Arrays to PyTorch Tensors**

```python
X_train_tensors = torch.from_numpy(X_train)
X_test_tensors = torch.from_numpy(X_test)
y_train_tensors = torch.from_numpy(y_train)
y_test_tensors = torch.from_numpy(y_test)
```

**Defining the model**

```python
from torch import float64

class MySimpleNeuralNetwork():
  def __init__(self,X):
    self.weights = torch.rand(X.shape[1],1,dtype = float64, requires_grad=True)  #i.e. 30 x 1 matrix
    self.bias = torch.zeros((1,1), dtype = float64, requires_grad=True)

  def forward(self,X):
    z = torch.matmul(X,self.weights) + self.bias
    y_pred = torch.sigmoid(z)
    return y_pred

  def loss(self, y_pred, y_true):
    epsilon = 1e-7
    y_pred = torch.clamp(y_pred, epsilon, 1-epsilon)

    loss = -(y_train_tensors * torch.log(y_pred) + (1-y_train_tensors) * torch.log(1-y_pred))
    return loss.mean()



```

```python
learning_rate = 0.1
epochs = 100
```

```python
# create model -> 3 steps (forward pass, loss calculate,bckward pass, parameters update) -> all in loop

model = MySimpleNeuralNetwork(X_train_tensors)

for epoch in range(epochs):
  #forward pass
  y_pred = model.forward(X_train_tensors)
  # print(y_pred)

  #loss calculate
  loss = model.loss(y_pred,y_train_tensors)

  #backward pass
  loss.backward()

  #parameters update
  with torch.no_grad():
    model.weights -= learning_rate * model.weights.grad
    model.bias -= learning_rate * model.bias.grad

  model.weights.grad.zero_()
  model.bias.grad.zero_()

  print(f'epoch: {epoch+1}, loss: {loss.item()}')
```
